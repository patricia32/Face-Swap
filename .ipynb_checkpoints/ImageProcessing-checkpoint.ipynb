```python
import numpy as np
import dlib
import time
import cv2

def extract_index_nparray(nparray):
    index = None
    for num in nparray[0]:
        index = num
        break
    return index


    # images reading  
image1 = cv2.imread("p2.jpg")
image1_gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
mask1 = np.zeros_like(image1_gray)

image2 = cv2.imread("p3.jpg")
image2_gray = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)
mask2 = np.zeros_like(image2_gray)



    # get detector and predictor
detector = dlib.get_frontal_face_detector()
predictor = dlib.shape_predictor("shape_predictor_68_face_landmarks.dat/shape_predictor_68_face_landmarks.dat")



    # result 1 settings
height, width, channels = image2.shape
img2_new_face = np.zeros((height, width, channels), np.uint8)

    # result 2 settings
height1, width1, channels1 = image1.shape
img1_new_face = np.zeros((height1, width1, channels1), np.uint8)



############## Face 1  ################
faces = detector(image1_gray)
for face in faces:
    landmarks = predictor(image1_gray, face)
    landmarks_points = []
    for n in range(0, 68):
        x = landmarks.part(n).x
        y = landmarks.part(n).y
        landmarks_points.append((x, y))


    points = np.array(landmarks_points, np.int32)
    convexhull = cv2.convexHull(points)
    cv2.fillConvexPoly(mask1, convexhull, 255)

    face_image_1 = cv2.bitwise_and(image1, image1, mask=mask1)

    # Delaunay triangulation
    rect = cv2.boundingRect(convexhull)
    subdiv = cv2.Subdiv2D(rect)
    subdiv.insert(landmarks_points)
    triangles = subdiv.getTriangleList()
    triangles = np.array(triangles, dtype=np.int32)

    indexes_triangles = []
    for t in triangles:
        pt1 = (t[0], t[1])
        pt2 = (t[2], t[3])
        pt3 = (t[4], t[5])

        index_pt1 = np.where((points == pt1).all(axis=1))
        index_pt1 = extract_index_nparray(index_pt1)

        index_pt2 = np.where((points == pt2).all(axis=1))
        index_pt2 = extract_index_nparray(index_pt2)

        index_pt3 = np.where((points == pt3).all(axis=1))
        index_pt3 = extract_index_nparray(index_pt3)

        if index_pt1 is not None and index_pt2 is not None and index_pt3 is not None:
            triangle = [index_pt1, index_pt2, index_pt3]
            indexes_triangles.append(triangle)

    ###############for second swap##############
    lines_space_mask2 = np.zeros_like(image2_gray)
    lines_space_new_face2 = np.zeros_like(image1)
    ############################################
    
###############################################################



#########################  Face 2  ############################
        
    faces2 = detector(image2_gray)
    for face in faces2:
        landmarks = predictor(image2_gray, face)
        landmarks_points2 = []
        for n in range(0, 68):
            x = landmarks.part(n).x
            y = landmarks.part(n).y
            landmarks_points2.append((x, y))


        points2 = np.array(landmarks_points2, np.int32)
        convexhull2 = cv2.convexHull(points2)
        
        ###############  for second swap  ##############
        cv2.fillConvexPoly(mask2, convexhull2, 255)

        face_image_2 = cv2.bitwise_and(image2, image2, mask=mask2)

        # Delaunay triangulation
        rect12 = cv2.boundingRect(convexhull2)
        subdiv2 = cv2.Subdiv2D(rect12)
        subdiv2.insert(landmarks_points2)
        triangles2 = subdiv2.getTriangleList()
        triangles2 = np.array(triangles2, dtype=np.int32)

        indexes_triangles2 = []
        for t in triangles2:
            pt12 = (t[0], t[1])
            pt22 = (t[2], t[3])
            pt32 = (t[4], t[5])

            index_pt12 = np.where((points2 == pt12).all(axis=1))
            index_pt12 = extract_index_nparray(index_pt12)

            index_pt22 = np.where((points2 == pt22).all(axis=1))
            index_pt22 = extract_index_nparray(index_pt22)

            index_pt32 = np.where((points2 == pt32).all(axis=1))
            index_pt32 = extract_index_nparray(index_pt32)

            if index_pt12 is not None and index_pt22 is not None and index_pt32 is not None:
                triangle2 = [index_pt12, index_pt22, index_pt32]
                indexes_triangles2.append(triangle2)
        ################################################
            
    lines_space_mask = np.zeros_like(image1_gray)
    lines_space_new_face = np.zeros_like(image2)
###############################################################
    
    
    
    
################## Triangulation of both faces ################
    for triangle_index in indexes_triangles:
       
        ########## Triangulation of the first face ##########
        tr1_pt1 = landmarks_points[triangle_index[0]]
        tr1_pt2 = landmarks_points[triangle_index[1]]
        tr1_pt3 = landmarks_points[triangle_index[2]]
        triangle1 = np.array([tr1_pt1, tr1_pt2, tr1_pt3], np.int32)


        rect1 = cv2.boundingRect(triangle1)
        (x, y, w, h) = rect1
        cropped_triangle = image1[y: y + h, x: x + w]
        cropped_tr1_mask = np.zeros((h, w), np.uint8)


        points = np.array([[tr1_pt1[0] - x, tr1_pt1[1] - y],
                           [tr1_pt2[0] - x, tr1_pt2[1] - y],
                           [tr1_pt3[0] - x, tr1_pt3[1] - y]], np.int32)

        cv2.fillConvexPoly(cropped_tr1_mask, points, 255)

        # Lines space   for first swap
        cv2.line(lines_space_mask, tr1_pt1, tr1_pt2, 255)
        cv2.line(lines_space_mask, tr1_pt2, tr1_pt3, 255)
        cv2.line(lines_space_mask, tr1_pt1, tr1_pt3, 255)
        lines_space = cv2.bitwise_and(image1, image1, mask=lines_space_mask)

        
        ########### Triangulation of second face ##########
        tr2_pt1 = landmarks_points2[triangle_index[0]]
        tr2_pt2 = landmarks_points2[triangle_index[1]]
        tr2_pt3 = landmarks_points2[triangle_index[2]]
        triangle2 = np.array([tr2_pt1, tr2_pt2, tr2_pt3], np.int32)


        rect2 = cv2.boundingRect(triangle2)
        (x2, y2, w2, h2) = rect2
        cropped_triangle2 = image2[y: y2 + h2, x: x2 + w2]  ## second
        cropped_tr2_mask = np.zeros((h2, w2), np.uint8)

        points2 = np.array([[tr2_pt1[0] - x2, tr2_pt1[1] - y2],
                            [tr2_pt2[0] - x2, tr2_pt2[1] - y2],
                            [tr2_pt3[0] - x2, tr2_pt3[1] - y2]], np.int32)

        cv2.fillConvexPoly(cropped_tr2_mask, points2, 255)

        
        # Warp triangles
        points = np.float32(points)
        points2 = np.float32(points2)
        M = cv2.getAffineTransform(points, points2)
        
        
        warped_triangle = cv2.warpAffine(cropped_triangle, M, (w2, h2))
        warped_triangle = cv2.bitwise_and(warped_triangle, warped_triangle, mask=cropped_tr2_mask)
        
        # Reconstructing destination face
        img2_new_face_rect_area = img2_new_face[y2: y2 + h2, x2: x2 + w2]
        img2_new_face_rect_area_gray = cv2.cvtColor(img2_new_face_rect_area, cv2.COLOR_BGR2GRAY)
        _, mask_triangles_designed = cv2.threshold(img2_new_face_rect_area_gray, 1, 255, cv2.THRESH_BINARY_INV)
        warped_triangle = cv2.bitwise_and(warped_triangle, warped_triangle, mask=mask_triangles_designed)

        img2_new_face_rect_area = cv2.add(img2_new_face_rect_area, warped_triangle)
        img2_new_face[y2: y2 + h2, x2: x2 + w2] = img2_new_face_rect_area
        
###############################################################

################## Triangulation of both faces ################   FOR SECOND SWAP
    for triangle_index in indexes_triangles2:
       
        ########## Triangulation of the first face ##########
        tr1_pt1 = landmarks_points2[triangle_index[0]]
        tr1_pt2 = landmarks_points2[triangle_index[1]]
        tr1_pt3 = landmarks_points2[triangle_index[2]]
        triangle1 = np.array([tr1_pt1, tr1_pt2, tr1_pt3], np.int32)


        rect1 = cv2.boundingRect(triangle1)
        (x, y, w, h) = rect1
        cropped_triangle = image2[y: y + h, x: x + w]
        cropped_tr1_mask = np.zeros((h, w), np.uint8)


        points = np.array([[tr1_pt1[0] - x, tr1_pt1[1] - y],
                           [tr1_pt2[0] - x, tr1_pt2[1] - y],
                           [tr1_pt3[0] - x, tr1_pt3[1] - y]], np.int32)

        cv2.fillConvexPoly(cropped_tr1_mask, points, 255)

        # Lines space   for first swap
        cv2.line(lines_space_mask2, tr1_pt1, tr1_pt2, 255)
        cv2.line(lines_space_mask2, tr1_pt2, tr1_pt3, 255)
        cv2.line(lines_space_mask2, tr1_pt1, tr1_pt3, 255)
        lines_space = cv2.bitwise_and(image2, image2, mask=lines_space_mask2)

        
        
        ########### Triangulation of second face ##########
        tr2_pt1 = landmarks_points[triangle_index[0]]
        tr2_pt2 = landmarks_points[triangle_index[1]]
        tr2_pt3 = landmarks_points[triangle_index[2]]
        triangle2 = np.array([tr2_pt1, tr2_pt2, tr2_pt3], np.int32)


        rect2 = cv2.boundingRect(triangle2)
        (x2, y2, w2, h2) = rect2
        cropped_triangle2 = image1[y: y2 + h2, x: x2 + w2]  ## second
        cropped_tr2_mask = np.zeros((h2, w2), np.uint8)

        points2 = np.array([[tr2_pt1[0] - x2, tr2_pt1[1] - y2],
                            [tr2_pt2[0] - x2, tr2_pt2[1] - y2],
                            [tr2_pt3[0] - x2, tr2_pt3[1] - y2]], np.int32)

        cv2.fillConvexPoly(cropped_tr2_mask, points2, 255)

        
        # Warp triangles
        points = np.float32(points)
        points2 = np.float32(points2)
        M = cv2.getAffineTransform(points, points2)
        
        warped_triangle = cv2.warpAffine(cropped_triangle, M, (w2, h2))
        warped_triangle = cv2.bitwise_and(warped_triangle, warped_triangle, mask=cropped_tr2_mask)
        

        # Reconstructing destination face
        img1_new_face_rect_area = img1_new_face[y2: y2 + h2, x2: x2 + w2]
        img1_new_face_rect_area_gray = cv2.cvtColor(img1_new_face_rect_area, cv2.COLOR_BGR2GRAY)
        _, mask_triangles_designed = cv2.threshold(img1_new_face_rect_area_gray, 1, 255, cv2.THRESH_BINARY_INV)
        warped_triangle = cv2.bitwise_and(warped_triangle, warped_triangle, mask=mask_triangles_designed)

        img1_new_face_rect_area = cv2.add(img1_new_face_rect_area, warped_triangle)
        img1_new_face[y2: y2 + h2, x2: x2 + w2] = img1_new_face_rect_area
        
         
###############################################################



# Face swapped (putting 1st face into 2nd face)
img2_face_mask = np.zeros_like(image2_gray)
img2_head_mask = cv2.fillConvexPoly(img2_face_mask, convexhull2, 255)
img2_face_mask = cv2.bitwise_not(img2_head_mask)



# Face swapped (putting 2nd face into 2st face)
img1_face_mask = np.zeros_like(image1_gray)
img1_head_mask = cv2.fillConvexPoly(img1_face_mask, convexhull, 255)
img1_face_mask = cv2.bitwise_not(img1_head_mask)


img2_head_noface = cv2.bitwise_and(image2, image2, mask=img2_face_mask)
img1_head_noface = cv2.bitwise_and(image1, image1, mask=img1_face_mask)
result = cv2.add(img2_head_noface, img2_new_face)
result2 = cv2.add(img1_head_noface, img1_new_face)

(x, y, w, h) = cv2.boundingRect(convexhull2)
center_face2 = (int((x + x + w) / 2), int((y + y + h) / 2))

seamlessclone = cv2.seamlessClone(result, image2, img2_head_mask, center_face2, cv2.NORMAL_CLONE)


###second
(x, y, w, h) = cv2.boundingRect(convexhull)
center_face1 = (int((x + x + w) / 2), int((y + y + h) / 2))

seamlessclone2 = cv2.seamlessClone(result2, image1, img1_head_mask, center_face1, cv2.NORMAL_CLONE)



dim = (400, 450)

resizedImg1 = cv2.resize(image1, dim, interpolation = cv2.INTER_AREA)
resizedImg2 = cv2.resize(image2, dim, interpolation = cv2.INTER_AREA)
resizedImgResult = cv2.resize(seamlessclone, dim, interpolation = cv2.INTER_AREA)
resizedImgResult2 = cv2.resize(seamlessclone2, dim, interpolation = cv2.INTER_AREA)

cv2.imshow("Img1", resizedImg1)
cv2.imshow("Img2", resizedImg2)
cv2.imshow("Img2 with face 1", resizedImgResult)
cv2.imshow("Img1 with face 2", resizedImgResult2)
cv2.waitKey(0)



cv2.destroyAllWindows()
```
